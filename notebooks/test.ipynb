```python
import dotenv
from agents import Agent, ModelSettings, Runner, function_tool, trace

dotenv.load_dotenv()
```

```python
@function_tool
def test_function(country : str) -> str : 
    """
        This function is to determine where you can have the best food for the country 
        Args : 
            country : country name
        returns
            liste of foods 
    """
    foods = {
        "France" : ["croissant", "boeuf bougignon"], 
        "Espagne" : ["paela", "Gasapachio"],
        "Ghana" : ["Igname", "Taro"], 
    }
    if country in foods : 
        return foods[country]
    else : 
        return "This country is not listed into my database"
```

```python
foods_agent = Agent(
    name = "food_agent", 
    instructions=""" You're a chef cooker, 
    List me the top two best food in each country I'll give to you
    Use the tools first and if you don't have any answer then use your own kwnoledge
""", 
tools=[test_function]

)
```

```python
with trace("foods advisor") : 
    result = await Runner.run(foods_agent, "Give me the top foods in Cote Ivoire")

    print(result.final_output)
```
